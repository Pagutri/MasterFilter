### Probando el m√≥dulo filtros.py

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import filtros
```

Probando el pasabajos:

```python
img_prueba = cv2.imread('imagenes/prueba.tif', 0)
```

```python
ker_ideal_pasabajos = filtros.kernel_pasabajos(img_prueba)
ker_gauss_pasabajos = filtros.kernel_pasabajos(img_prueba, forma = 1)
ker_btw_pasabajos = filtros.kernel_pasabajos(img_prueba, forma = 2)
```

```python
prueba_ideal_pasabajos = filtros.aplicar_filtro(img_prueba, ker_ideal_pasabajos)
prueba_gauss_pasabajos = filtros.aplicar_filtro(img_prueba, ker_gauss_pasabajos)
prueba_btw_pasabajos = filtros.aplicar_filtro(img_prueba, ker_btw_pasabajos)
```

```python
fig = plt.figure(figsize = (15, 10))
fig.add_subplot(1, 4, 1)
plt.imshow(img_prueba, cmap = 'gray')
plt.title('Imagen original', size = 18);
fig.add_subplot(1, 4, 2)
plt.imshow(prueba_ideal_pasabajos, cmap = 'gray')
plt.title('Pasabajos ideal', size = 18);
fig.add_subplot(1, 4, 3)
plt.imshow(prueba_gauss_pasabajos, cmap = 'gray')
plt.title('Pasabajos gaussiano', size = 18);
fig.add_subplot(1, 4, 4)
plt.imshow(prueba_btw_pasabajos, cmap = 'gray')
plt.title('Pasabajos de Butterworth', size = 18);
```
